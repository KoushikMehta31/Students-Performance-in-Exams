```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10, 6)
sns.set(style="whitegrid")

df.to_csv("Cleaned_StudentsPerformance.csv", index=False)
print("Cleaned data saved")
```

```python
sns.countplot(x='gender',color='steelblue', edgecolor='black', data=df)
plt.title("Student Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.countplot(x='test preparation course',color='steelblue', edgecolor='black', data=df)
plt.title("Test Preparation Completion")
plt.xlabel("Test Preparation Course")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.countplot(x='test preparation course', hue='gender', data=df, edgecolor='black')

plt.title("Test Preparation Completion by Gender")
plt.xlabel("Test Preparation Course")
plt.ylabel("Number of Students")
plt.legend(title='Gender', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x='gender', y='math score', data=df)
plt.title("Math Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()
```

```python
sns.boxplot(x='gender', y='writing score', data=df)
plt.title("Writing Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Writing Score")
plt.show()
```

```python
sns.boxplot(x='gender', y='reading score', data=df)
plt.title("Reading Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Reading Score")
plt.show()
```

```python
avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

avg_scores.plot(kind='bar', color=['blue', 'orange', 'green'],edgecolor='black')
plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.legend(title='Subjects', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
race_counts = df['race/ethnicity'].value_counts()

plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%')
plt.title("Race/Ethnicity Distribution")
plt.axis('equal')
plt.show()
```

```python
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Scores")
plt.xlabel("Subjects")
plt.ylabel("Subjects")
plt.show()
```

```python
sns.violinplot(x='test preparation course', y='reading score', data=df)
plt.title("Reading Score by Test Preparation")
plt.xlabel("Test Preparation Course")
plt.ylabel("Reading Score")
plt.show()
```

```python
sns.violinplot(x='test preparation course', y='writing score', data=df)
plt.title("Writing Score by Test Preparation")
plt.xlabel("Test Preparation Course")
plt.ylabel("Writing Score")
plt.show()
```

```python
avg_scores = df.groupby('test preparation course')[['reading score', 'writing score']].mean()

avg_scores.plot(kind='bar', color=['skyblue', 'salmon'],edgecolor='black')
plt.title("Average Reading & Writing Scores by Test Preparation Course")
plt.xlabel("Test Preparation Course")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.legend(title='Subject')
plt.tight_layout()
plt.show()
```

```python
sns.stripplot(x='parental level of education', y='math score', data=df, jitter=True)

plt.xticks(rotation=45)
plt.title("Math Score vs Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Math Score")
plt.tight_layout()
plt.show()
```

```python
sns.stripplot(x='parental level of education', y='reading score', data=df, jitter=True)

plt.xticks(rotation=45)
plt.title("Reading Score vs Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Reading Score")
plt.tight_layout()
plt.show()
```

```python
sns.stripplot(x='parental level of education', y='writing score', data=df, jitter=True)

plt.xticks(rotation=45)
plt.title("Writing Score vs Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Writing Score")
plt.tight_layout()
plt.show()
```

```python
avg_scores = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()

avg_scores.plot(kind='bar', color=['blue', 'orange', 'green'], edgecolor='black')

plt.xticks(rotation=45)
plt.title("Average Scores by Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Average Score")
plt.legend(title='Subjects', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
